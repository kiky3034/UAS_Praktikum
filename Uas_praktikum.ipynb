```python
import pandas as pd
import matplotlib.pyplot as plt

# Membaca data dari file CSV
data = pd.read_csv('stock.csv')

# Mengambil kolom "Purchase Price", "Current Price", dan "Gain/Loss"
purchase_price = data['Purchase Price']
current_price = data['Current Price']
gain_loss = data['Gain/Loss']

# Membuat histogram Purchase Price
plt.hist(purchase_price, bins=10, color='blue')
plt.xlabel('Purchase Price')
plt.ylabel('Frequency')
plt.title('Distribution of Purchase Price')
plt.show()

# Membuat histogram Current Price
plt.hist(current_price, bins=10, color='green')
plt.xlabel('Current Price')
plt.ylabel('Frequency')
plt.title('Distribution of Current Price')
plt.show()

# Membuat histogram Gain/Loss
plt.hist(gain_loss, bins=10, color='red')
plt.xlabel('Gain/Loss')
plt.ylabel('Frequency')
plt.title('Distribution of Gain/Loss')
plt.show()
```
